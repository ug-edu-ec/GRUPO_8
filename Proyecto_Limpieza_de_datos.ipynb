##Transformar los años a filas

```python
def get_years():
  years = []
  for i in range(2015,2021):
    years.append(str(i))
  return years
```

```python
dfa = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
               ,value_vars=get_years(), var_name='Year', value_name='Valor')
dfa['Year'] = dfa['Year'].astype(int)
dfa.sample(5)
```
